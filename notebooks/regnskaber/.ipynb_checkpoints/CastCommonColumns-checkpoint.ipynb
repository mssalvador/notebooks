```python
def getNumberOfTypes(df,broadcastCount):
    '''
        The method returns the distinct values for each column

        Input:
            df - Data Frame with x columns

        Output:
            output - tuple with list as elements the inner elements are the different types in the column
    '''
    output = {}
    for c in df.columns:
        types = df.select(F.col(c)).filter(F.col(c) != " ").count()
        output[c] = float(types)/float(broadcastCount.value) 
    sqlContext.clearCache()
    return output
```

```python
def checkTypes(val):
    '''
        hello world
        
        Input:
            va - val a column
        
        Output:
            output - data frame with casted types. 
    '''
    if val == None: return (val,"BLANK")
    if val == " ": return (None,"BLANK")
    wordRe = re.compile(r'(\w\s\w)+',re.I | re.LOCALE)
    dateRe = re.compile(r'\d{4}\W\d{2}\W\d{2}|\d{2}\W\d{2}\W\d{4}',re.I | re.LOCALE)
    valueRe = re.compile(r'[^a-zA-Z\s]+') #should only get numbers and symbols back e.g.: 100.000,00 or 2014-01-01
    try:
        valRe = valueRe.search(val)
        if valRe:
            dateVal = dateRe.match(valRe.group(0))
            if dateVal:
                if dateVal.group(0) == val:
                    if re.match(r"\d{4}\W\d{2}\W\d{2}",val):
                        return (datetime.strptime(val,"%Y-%m-%d").date(),"DATE")
                    else:
                        return (datetime.strptime(val,"%d-%m-%Y").date(),"DATE")
            elif valRe.group(0) == val:
                tVal = re.sub(",",".",re.sub(r"\.","",val))
                #CheckForIntvals = re.match("[\d+\-\d+]+",tVal)
                #if CheckForIntvals:
                #    if CheckForIntvals.group(0) == tVal:
                #        return (tVal,"TEXT")
                return (float(tVal),"FLOAT")
        
    except AttributeError:
        print("Something's not right: ", str(val))
        return val
    except ValueError:
        
        print("ValRe: "+str(valRe.group(0))+" Val: "+str(val))
        return (tVal,"TEXT")
    else:
        return (val,"TEXT")
```

```python
def checkRow(rdd):
    return [checkTypes(x) for x in rdd]
```

```python
def reportRow(rdd):
    
    return [checkTypes(x) for x in rdd ]
```

```python
def IdentType(val):
    if val == "TEXT":
        return StringType()
    elif val == "DATE":
        return DateType()
    else:
        return DoubleType()
```
